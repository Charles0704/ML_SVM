```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense 
from keras.optimizers import gradient_descent_v2 
```

```python
data=pd.read_csv("ex2data1.txt",names=["Exam_1","Exam_2","Accepted"])
data.head(10)
```

```python
x=data.iloc[:, :-1]
y=data.iloc[:,-1]
```

```python
model = Sequential()
model.add(tf.keras.layers.Dense(1,input_shape=(2,))) #inputlayer
model.add(Dense(units=50,input_dim=1,activation='relu'))  #hidden layer
model.add(Dense(units=150,input_dim=1,activation='relu'))  #hidden layer
model.add(Dense(units=150,input_dim=1,activation='relu'))  #hidden layer
model.add(Dense(units=50,input_dim=1,activation='relu'))  #hidden layer
model.add(Dense(units=1,activation='sigmoid'))  #output layer
model.summary()
```

```python
model.compile(optimizer='adam',loss = 'binary_crossentropy',metrics=['acc'])
```

```python
history=model.fit(x,y,epochs=500)
```

```python
history_df = pd.DataFrame(history.history)
history_df['loss'].plot();
history_df['acc'].plot();
plt.tick_params(colors='white')
plt.tick_params(colors='white')
```

```python
model.predict(data[1:len(data)].iloc[:,:-1])
result=list(range(1,len(data)+1))
for i in range(1,len(data)):
    if(model.predict(data[:len(data)].iloc[:,:-1])[i])<0.5:
        result[i]=0
    else:
        result[i]=1    
```

```python
X=list(range(1,len(data)+1))
sum=0
for j in range(1,len(data)):
    if (result[j]!=list(y)[j]):
        X[j]=0
    else: X[j]=1
    sum+=X[j]  
print("Accuracy rate: ",sum/len(data))       
```
