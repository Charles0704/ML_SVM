# SVM

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC
from sklearn import svm
from sklearn.svm import SVC
from sklearn import metrics 
```

```python
data=pd.read_csv("ex2data1.txt",names=["Exam_1","Exam_2","Accepted"])
data2=pd.read_csv("ex2data2.txt",names=["Test_1","Test_2","Accepted"])
```

```python
X=(data.iloc[:,0:-1]).values
Y=data.iloc[:,-1].values.reshape(len(data.iloc[:,-1]),-1)
clf=svm.SVC(kernel='linear')
clf.fit(X,Y)
x=np.linspace(20,110,100)
w=clf.coef_[0]
a=-w[0]/w[1]
y_svm=a*x-(clf.intercept_[0])/w[1]
x=np.linspace(20,110,100)
```

```python
plt.figure(figsize=(10, 8))
plt.plot(x,y_svm)
plt.scatter(list((data[data['Accepted']==0]["Exam_1"]).values),list((data[data['Accepted']==0]["Exam_2"]).values),color='red',marker='o',label='y=0')
plt.scatter(list((data[data['Accepted']==1]["Exam_1"]).values),list((data[data['Accepted']==1]["Exam_2"]).values),color='blue',marker='o',label='y=0')
plt.xlim(20, 110)
plt.ylim(20, 110)
plt.legend(loc="upper right")
plt.xlabel("Exam1")
plt.ylabel("Exam2")
```

```python
X=(data.iloc[:,0:-1]).values
Y=data.iloc[:,-1].values.reshape(len(data.iloc[:,-1]),-1)
plt.figure(figsize=(10, 8))
poly_kernel_svm=Pipeline((
    ("scaler",StandardScaler()),
    ("svm_clf",SVC(kernel="poly",degree=5,coef0=2,C=50))
))
poly_kernel_svm.fit(X,Y)
xx, yy = np.meshgrid(np.arange(20,110,0.05), np.arange(20,110,0.05))
y_new=poly_kernel_svm.predict(np.c_[xx.ravel(),yy.ravel()])
plt.contourf(xx, yy, y_new.reshape(xx.shape),cmap="PuBu")
plt.scatter(X[:,0],X[:,1],marker="o",c=Y)
plt.axis([20,110,20,110])
print("Accuracy: ",metrics.accuracy_score(Y, poly_kernel_svm.predict(X)))
```

```python
plt.figure(figsize=(10, 8))
rbf_kernel_svm=Pipeline((
    ("scaler",StandardScaler()),
    ("svm_clf",SVC(kernel="rbf",gamma=1))
))
rbf_kernel_svm.fit(X,Y)
xx, yy = np.meshgrid(np.arange(20,110,0.05), np.arange(20,110,0.05))
y_new=rbf_kernel_svm.predict(np.c_[xx.ravel(),yy.ravel()])
plt.contourf(xx, yy, y_new.reshape(xx.shape),cmap="YlOrRd")
plt.scatter(X[:,0],X[:,1],marker="o",c=Y)
plt.axis([20,110,20,110])
print("Accuracy: ",metrics.accuracy_score(Y, rbf_kernel_svm.predict(X)))
```
